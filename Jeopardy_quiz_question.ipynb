```python
import re
# lowercase chars, remove punctuation
def cleaner(column):
    s = column.lower()
    s = re.sub(r'[^A-Za-z0-9\s]', '', s)
    return s

jeopardy["clean_answer"] = jeopardy["Answer"].apply(cleaner)
jeopardy["clean_question"] = jeopardy["Question"].apply(cleaner)


jeopardy["clean_question"]
jeopardy["clean_answer"]
```

```python
# convert numeric values to int
# error = 0
def cleaner2(column):
    s = column.lower()
    s = re.sub(r'[^A-Za-z0-9\s]', '', s)
    try:
        s = int(s)
    except Exception:
        s = 0
    return s

jeopardy["clean_value"] = jeopardy["Value"].apply(cleaner2)
jeopardy["clean_value"]

# convert column with dates to date format
jeopardy["clean_Air"] = pd.to_datetime(jeopardy["Air Date"])
jeopardy["clean_Air"]
```

## Check if answer appear in question

```python
def spliter(row):
    split_question = row["clean_question"].split(" ")
    split_answer = row["clean_answer"].split(" ")
    
    match_count = 0
    
    if "the" in split_answer:
        split_answer.remove("the")
        
    if len(split_answer) == 0:
        return 0
    
    for r in split_answer:
        if r in split_question:
            match_count += 1
    return match_count / len(split_answer)

jeopardy["answer_in_question"] = jeopardy.apply(spliter, axis=1)
jeopardy["answer_in_question"].mean()
```

There is only ~ 6% of questions where answer appears. It is not huge value, hence if you would like to participate in Jeopardy you have to be prepared (study).
